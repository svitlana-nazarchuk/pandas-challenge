```python
# Dependencies and Setup
import pandas as pd

#import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"


# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)


student_data = pd.read_csv(student_data_to_load)


#Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data,  how="left", on="school_name")
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete["school_name"].unique())

total_students = len(school_data_complete["student_name"])
total_students_i = format(total_students, ",d")


total_budget = school_data_complete["budget"].unique()
total_budget = total_budget.sum()
#total_budget = format(total_budget, ",d")


avg_math_score = school_data_complete["math_score"].mean()
avg_math_score = format(avg_math_score, ".2f" )


avg_reading_score = school_data_complete["reading_score"].mean()
avg_reading_score = format(avg_reading_score, ".2f")



passing_math_students = len(school_data_complete.loc[school_data_complete["math_score"]>69])/(total_students)
passing_math_students_f = format(passing_math_students, ".2%")

passing_reading_students = len(school_data_complete.loc[school_data_complete["reading_score"]>69])/total_students
passing_reading_students_f = format(passing_reading_students, ".2%")

passing_rate = (float(passing_math_students) + float(passing_reading_students))/2
passing_rate = format(passing_rate, ".2%")


district_summary = pd.DataFrame({"Total Schools": total_schools,
                                 "Total Students": total_students_i,
                                 "Total Budget": total_budget,
                                 "Average Math Score": avg_math_score,
                                 "Average Reading Score": avg_reading_score,
                                 "Passing Math": passing_math_students_f,
                                 "Passing Reading": passing_reading_students_f,
                                 "Overall Passing Rate": [passing_rate]
                                })

district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```
